### Law of Large Numbers (coin toss)

```python
import numpy as np
import pandas as pd 
import seaborn as sns

def coin_flip(nr_flips) -> pd.DataFrame:
    # Geneate random numbers and use decision threshold 0.5 for heads vs tail
    n = [np.random.uniform(0,1) for i in range(nr_flips)]
    heads = [1 if i > 0.5 else 0 for i in n]
    proportion = np.cumsum(heads)/np.arange(1, nr_flips + 1)
    print(heads)
    
    df = pd.DataFrame(
        {"flips": np.arange(1, nr_flips + 1),
        "heads": np.cumsum(heads), 
        "proportion_heads": proportion
    })
    
    return df 
    
```

```python
flips = 1000

nr_runs = 5

df = pd.DataFrame()
for i in range(nr_runs):
    run = coin_flip(flips)
    run['runs'] = "run{}".format(i+1)
    df = pd.concat([df, run], axis = 0)
display(df)
```

```python
ax = sns.lineplot(data = df, x = "flips", y = "proportion_heads", hue = "runs")
```
